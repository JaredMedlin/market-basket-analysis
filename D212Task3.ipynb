```python
#Dropping rows with no entries
df2=df[df['Item01'].notna()]
```

```python
#Transforming Data w/ TransactionEncoder
#Code sourced from either Dr. Kamara's videos or
#https://rasbt.github.io/mlxtend/USER_GUIDE_INDEX/
from mlxtend.preprocessing import TransactionEncoder
#Convert to lists
rows = []
for i in range (0, 7501):
    rows.append([str(df2.values[i,j])
for j in range (0, 20)])
```

```python
#TransactionEncoder from mlxtend
DE = TransactionEncoder()
array = DE.fit(rows).transform(rows)
transaction = pd.DataFrame(array, columns = DE.columns_)
transaction
```

```python
#List of columns
for col in transaction.columns:
    print(col)
```

```python
#Checking for empty columns and removing
df4 = transaction.drop(['nan'], axis = 1)
df4.head(7505)
```

```python
df4.to_csv('C:\\Users\\weepo\\Documents\\WGU\\D212\\task 3\\teleco_market_basket_cleaned.csv')
```

```python
#Apriori time
#Code sourced from either Dr. Kamara's videos or
#https://rasbt.github.io/mlxtend/USER_GUIDE_INDEX/
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
rules = apriori(df4, min_support = 0.02, use_colnames = True)
rules.head(5)
```

```python
#Rules table
rulestable = association_rules(rules, metric = 'lift', min_threshold = 1)
rulestable.head(10)
```

```python
#following code is for showing top 3 rules for each metric
top3 = rulestable.sort_values('confidence', ascending=False).head(3)
top3
```

```python
top3 = rulestable.sort_values('lift', ascending=False).head(3)
top3
```

```python
top3 = rulestable.sort_values('support', ascending=False).head(3)
top3
```
